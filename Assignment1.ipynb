5.What are the advantages and disadvantages of a very flexible (versus a less flexible) approach for regression or classification? Under what circumstances might a more flexible approach be preferred to a less flexible approach? When might a less flexible approach be preferred?<br><br>
The advantage of a flexible approach is that flexible models can fit multiple different shapes for f. The downside is that there is higher potential for overfitting the data. A flexible approach is preferred when you’re more interested in prediction (as opposed to inference). A less flexible approach is preferred if your goal is inference and interpretability. More flexible approaches are often more complicated and less easy to interpret and understand.  


6.Describe the differences between a parametric and a non-parametric statistical learning approach. What are the advantages of a para- metric approach to regression or classification (as opposed to a non- parametric approach)? What are its disadvantages? <br><br>
The main difference between parametric and non-parametric approaches are assumptions. Parametric tests must follow a certain set of assumptions, while non-parametric tests do not. Since non-parametric tests are less rigid, they can accurately fit a wider range of shapes for f. The advantage of a parametric test is that it is simpler – when assumptions are met, you know what you’re getting and you don’t need as many observations. The disadvantage of parametric tests happens when your data does not fit the shape that the test takes (for example linear), resulting in inaccurate estimates of the data.
